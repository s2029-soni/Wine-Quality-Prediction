## **Define Target Variable (y) and Feature Variables (X)**


```python
y=df['quality']
y.shape
```

```python
y
```

```python
x=df.drop(['quality'],axis=1)
x.shape
```

```python
x
```

## **Data Preprocessing**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
x
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3,stratify=y, random_state=2259)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## **Modeling**

```python
from sklearn.svm import SVC
svc =SVC()
svc.fit(x_train,y_train)
```

## **Prediction**

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
y_pred.shape
```

## **Model Evaluation**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

# **Getting Model rerun with Two Class Created for wine quality**

Wine quality 3,4,5 labelled as 0

Wine quality 6,7,8,9 labelled as 1


```python
y=df['quality'].apply(lambda y_value:1 if y_value>=6 else 0 )
```

```python
y.value_counts()
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3,stratify=y, random_state=2259)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## **Modeling**

```python
from sklearn.svm import SVC
svc =SVC()
svc.fit(x_train,y_train)
```

## **Prediction**

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
y_pred.shape
```

## **Model evaluation**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
